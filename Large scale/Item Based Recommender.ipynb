```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
```

```python
header = ['user_id','item_id','rating','timestamp']
dataset = pd.read_csv('C:\\Users\\User\\Desktop\\Data_Science\\Large scale\\Items.csv',names = header)
```

```python
dataset.head()
```

```python
Mean = dataset.groupby(by="item_id",as_index=False)['rating'].mean()

Rating_avg = pd.merge(Mean,dataset, on='item_id')
Rating_avg.columns=['item_id', 'rating_item','user_id', 'avg_rating','timestamp']
Rating_avg['adg_rating']=Rating_avg['rating_item']-Rating_avg['avg_rating']
Rating_avg.head()
```

```python
check = pd.pivot_table(Rating_avg,values='rating_item',index='item_id',columns='user_id')
check.head()
```

```python
final = pd.pivot_table(Rating_avg,values='adg_rating',index='item_id',columns='user_id')
final.head()
```

```python
final_item = final.apply(lambda row: row.fillna(row.mean()), axis=1)
final_item.head()
```

```python
final_user = final.fillna(final.mean(axis=0))
final_user.head()
```

```python
b = cosine_similarity(final_item)
np.fill_diagonal(b, 0 )
similarity_with_item = pd.DataFrame(b,index=final_item.index)
similarity_with_item.columns=final_item.index
similarity_with_item.head()
```

```python
# user similarity on replacing NAN by user avg
cosine = cosine_similarity(final_user)
np.fill_diagonal(cosine, 0)
similarity_with_user = pd.DataFrame(cosine, index=final_user.index)
similarity_with_user.columns = final_item.index
similarity_with_user.head()
```

```python
def find_n_neighbours(df,n):
    #order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```python
# top 30 neighbours for each item
sim_item_30 = find_n_neighbours(similarity_with_item,30)
sim_item_30.head()
```

```python
# top 30 neighbours for each user
sim_user_30 = find_n_neighbours(similarity_with_user,30)
sim_user_30.head()
```

```python
def get_similar_items( item1, item2 ):
    common = Rating_avg[Rating_avg.item_id == item1].merge(
    Rating_avg[Rating_avg.item_id == item2],
    on = "user_id",
    how = "inner" )
    return common.merge( dataset, on = 'user_id' )
```

```python
a = get_similar_items(242,302)
#a = a.loc[ : , ['rating_item_x','rating_item_y']]
a.head(10)
```

```python
def item_user_score(user,item):
    a = sim_user_30[sim_user_30.index==item].values # users, who rated the item
    b = a.squeeze().tolist()
    c = final_user.loc[:,user] # ratings given for items by the user
    d = c[c.index.isin(b)]
    f = d[d.notnull()]
    avg_item = Mean.loc[Mean['item_id'] == item,'rating'].values[0] 
    index = f.index.values.squeeze().tolist()
    corr = similarity_with_user.loc[item,index]
    fin = pd.concat([f, corr], axis=1)
    fin.columns = ['adg_score','correlation']
    fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
    nume = fin['score'].sum()
    deno = fin['correlation'].sum()
    final_score = avg_item + (nume/deno)
    return final_score
```

```python
score = item_user_score(242,302)
print("score (i, u) is",score)
```

```python
Rating_avg = Rating_avg.astype({"user_id": str})
User_item = Rating_avg.groupby(by = 'item_id')['user_id'].apply(lambda x:','.join(x))
User_item.index = User_item.index.astype('int')
User_item.head()
```

```python
def Item_user_score1(item):
    Item_seen_by_user = check.columns[check[check.index == item].notna(
    ).any()].tolist() # getting all users who's rated the item
    a = sim_user_30[sim_user_30.index == item].values # getting top 30 users who has been similar for the itrm
    b = a.squeeze().tolist()
    d = User_item[User_item.index.isin(b)]
    l = ','.join(d.values)
    Item_seen_by_similar_users = l.split(',')
    Item_under_consideration = list(
        set(Item_seen_by_similar_users)-set(list(map(str, Item_seen_by_user))))
    Item_under_consideration = list(map(int, Item_under_consideration))
    score = []
    for user in Item_under_consideration:
        c = final_user.loc[:, user]
        d = c[c.index.isin(b)]
        f = d[d.notnull()]
        avg_item = Mean.loc[Mean['item_id'] == item, 'rating'].values[0]
        index = f.index.values.squeeze().tolist()
        corr = similarity_with_user.loc[item, index]
        fin = pd.concat([f, corr], axis=1)
        fin.columns = ['adg_score', 'correlation']
        fin['score'] = fin.apply(
            lambda x: x['adg_score'] * x['correlation'], axis=1)
        nume = fin['score'].sum()
        deno = fin['correlation'].sum()
        final_score = avg_item + (nume/deno)
        score.append(final_score)
    data = pd.DataFrame({'user_id': Item_under_consideration, 'score': score})
    top_5_recommendation = data.sort_values(
        by='score', ascending=False).head(5)
    return top_5_recommendation
```

```python
item = int(input("Enter the item id that you want to recommend : "))
predicted_item = Item_user_score1(item)
print(" ")
print(f"The Recommendations for item id", item)
print("   ")
for i in range(5):
    print(predicted_item.iloc[i])
```
